```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 5, 10])
y = np.array([10, 15, 20])

x_new = 7.2
y_new = np.interp(x_new, x, y)

plt.plot(x, y, 'bo-', label='Data Kecepatan')

plt.plot(x_new, y_new, 'ro', label=f'Interpolasi ({x_new} m, {y_new:.2f} m/detik)')

plt.xlabel('Jarak (meter)')
plt.ylabel('Kecepatan (m/detik)')
plt.title('Interpolasi Linier Kecepatan Lari')

plt.grid(True)
plt.legend()

plt.show()

print(f"Kecepatan pada jarak {x_new} meter adalah {y_new:.2f} m/detik")
```
