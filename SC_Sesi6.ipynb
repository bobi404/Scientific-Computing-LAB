```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda t, s: np.exp(-t)

h = 0.1
t_min = 0
t_max = 1

t = np.arange(t_min, t_max + h, h)
s0 = -1

s = np.zeros(len(t))
s[0] = s0

for i in range(0, len(t) - 1):
    s[i + 1] = s[i] + h * f(t[i], s[i])

plt.figure(figsize=(12,8))
plt.plot(t, s, 'bo--', label='approx (euler h=0.1)')
plt.plot(t, -np.exp(-t), 'g', label='exact solution')
plt.xlabel('t')
plt.ylabel('s')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda x, y: x**2 - 4*y

h = 0.1
x_min = 0
x_max = 1

x = np.arange(x_min, x_max + h, h)
y0 = 1
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

exact = lambda x: (1/4)*x**2 - (1/8)*x + (31/128)*np.exp(-4*x)

plt.figure(figsize=(12,8))
plt.plot(x, y, 'bo--', label='approx (euler h=0.1)')
plt.plot(x, exact(x), 'g', label='exact solution')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Euler's Method vs Exact Solution")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-poster')

f = lambda x, y: np.sin(y)

h = 0.1
x_min = 0
x_max = 1

x = np.arange(x_min, x_max + h, h)
y0 = 1
y = np.zeros(len(x))
y[0] = y0

for i in range(0, len(x) - 1):
    y[i + 1] = y[i] + h * f(x[i], y[i])

plt.figure(figsize=(12,8))
plt.plot(x, y, 'bo--', label="approx (Euler's method, h=0.1)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("Euler's Method for y' = sin(y), y(0) = 1")
plt.grid(True)
plt.show()
```
